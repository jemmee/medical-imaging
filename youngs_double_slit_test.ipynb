```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Grid setup
width, height = 500, 500
x = np.linspace(-10, 10, width)
y = np.linspace(-10, 10, height)
X, Y = np.meshgrid(x, y)

# 2. Parameters (Wavelength and Slit distance)
wavelength = 1.5
d = 4.0  # Distance between slits

# 3. Calculate distance from each slit to every point in the grid
# Slit 1 at (-d/2, 0), Slit 2 at (d/2, 0)
r1 = np.sqrt((X + d/2)**2 + Y**2)
r2 = np.sqrt((X - d/2)**2 + Y**2)

# 4. Wave Equation: Amplitude = cos(2 * pi * r / lambda)
wave1 = np.cos(2 * np.pi * r1 / wavelength)
wave2 = np.cos(2 * np.pi * r2 / wavelength)

# 5. Interferometry: The Sum of waves
interference = wave1 + wave2

# 6. Plotting
plt.figure(figsize=(8, 8))
plt.imshow(interference, extent=[-10, 10, -10, 10], cmap='RdBu')
plt.title("Interferometry Simulation: Double Slit Interference")
plt.xlabel("Distance")
plt.ylabel("Distance")
plt.colorbar(label="Wave Amplitude")
plt.show()
```
