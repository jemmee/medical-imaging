```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Setup Time
t = np.linspace(0, 3000, 1000) # 0 to 3000 milliseconds

# 2. Define Relaxation Constants (in ms)
# T1 is the time it takes for protons to realign with the magnet
t1_fat = 250    # Fat realigns quickly
t1_csf = 2000   # Water/Cerebrospinal Fluid realigns slowly

# 3. Simulate T1 Recovery Curves
# Formula: Signal = 1 - exp(-t / T1)
signal_fat = 1 - np.exp(-t / t1_fat)
signal_csf = 1 - np.exp(-t / t1_csf)

# 4. Visualization
plt.figure(figsize=(10, 6))

plt.plot(t, signal_fat, label="Fat (Short T1)", color='orange', linewidth=2)
plt.plot(t, signal_csf, label="CSF/Water (Long T1)", color='blue', linewidth=2)

# Point out the "Contrast"
# If we take a picture at 500ms, notice the huge gap between Fat and CSF
plt.axvline(x=500, color='red', linestyle='--', label='Inversion Time (Capture)')

plt.title("MRI Physics: T1 Relaxation Curves")
plt.xlabel("Time after RF Pulse (ms)")
plt.ylabel("Signal Intensity")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
